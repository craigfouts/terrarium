```python
import matplotlib.pyplot as plt
import numpy as np
from sympy import solve, solveset, exp
from sympy.abc import t, x
from tqdm import tqdm
```

```python
def simulate_lde(r, seed=None, n_steps=10, verbosity=1):
    out = np.ndarray(n_steps)
    out[0] = n = seed

    for i in tqdm(range(1, n_steps)) if verbosity > 0 else range(1, n_steps):
        # out[i] = (n := n*r)  # de
        out[i] = out[i - 1]*np.exp(r*(1 - out[i - 1]))

    return out
```

```python
r = 2.1
seed = .9
n_steps = 100

lde = simulate_lde(r, seed, n_steps)
plt.plot(range(n_steps), lde)
plt.show()
```

```python
e2_args = (-.2, -9.8)
e3_args = (.5, 450)

def ode(y, a=1., b=1.):
    return a*y - b

t_range = np.arange(10)
e2_range = np.linspace(0, 100, 15)
e3_range = np.linspace(700, 1100, 15)

c_range = np.arange(-2, 3)

for c in c_range:
    plt.plot(t_range, c*np.exp(t_range/2) + 900, c='blue')

for t in tqdm(t_range):
    for y in e3_range:
        out = ode(y, *e3_args)
        plt.plot((t - .1, t + .1), (y - out*.2, y + out*.2), c='red')

plt.show()
```
